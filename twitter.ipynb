Comenzaremos por limipar la data y organizar la informacion en un dataset de pandas, con tal de que esta sea mas lejible y trabajable. Utilizaremos la metodologia que se utiliza en este lnk: https://www.kaggle.com/code/prathamsharma123/clean-raw-json-tweets-data

```python
import pandas as pd
from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings("ignore")
```

```python
#no se subio el archivo json a github debido a su tama√±o, para que el codigo corra es necesario este archivo. 
raw_tweets = pd.read_json('farmers-protest-tweets-2021-03-5.json', lines=True)
raw_tweets.head(5)
```

Luego notamos que el campo de user tiene un json en vez de texto, por lo que lo normalizaremos con pandas. 

```python
users = json_normalize(raw_tweets['user'])
users.drop(['description', 'linkTcourl'], axis=1, inplace=True)
users.rename(columns={'id':'userId', 'url':'profileUrl'}, inplace=True)
users.head(5)
```

Luego quitamos los duplicados

```python
users = pd.DataFrame(users)
users.drop_duplicates(subset=['userId'], inplace=True)
print("Shape: ", users.shape)
users.head(5)
```

Ahora podemos agregarle a nuestro dataframe de tweets el campo de username

```python
user_name = []
for user in raw_tweets['user']:
    una = user['username']
    user_name.append(una)
raw_tweets['username'] = user_name
```

```python
raw_tweets.head()
```

Continuamos quitando las columnas que no son de nuestro interes y modificando la columna de date para que pase de ser tipo datetime a date.

```python
cols = [ 'id', 'username', 'date', 'renderedContent', 'retweetCount']
tweets = raw_tweets[cols]
tweets['date'] = pd.to_datetime(tweets['date']).dt.date
tweets.rename(columns={'id':'tweetId'}, inplace=True)
tweets.head()
```

Ahora tenemos un dataframe de tweets que nos sirve para realizar los analisis pedidos. El siguiente paso es realizar las funciones que se nos piden. En todas retornaremos la informacion pedida como un dataframe con los datos relevantes de cada consuta.

```python
def top_retweet():
    sorted_df = tweets.sort_values(by=['retweetCount'], ascending=False)
    cols = ['renderedContent', 'retweetCount']
    top_retweets = sorted_df[cols][:10]
    print(top_retweets)
```

```python
def top_tweet():
    sorted_tweets = tweets['username'].value_counts()
    top_tweets = sorted_tweets[:10].to_frame().reset_index()
    top = top_tweets.rename(columns={"index": "userName", "username": "tweetCount"})
    print(top)
```

```python
def top_day():
    sorted_days = tweets['date'].value_counts()
    top_days = sorted_days[:10].to_frame().reset_index()
    top_d = top_days.rename(columns={"index": "Date", "date": "dateCount"})
    print(top_d)
```

```python
import re

def f(s, pat):
    pat = r'(\w*%s\w*)' % pat
    return re.findall(pat, s)

def top_hashtag():
    hashtags = {}
    # a = 0
    for index, row in tweets.iterrows():
    
        hashtag = f(row["renderedContent"], "#")
        for i in hashtag:
            if i in hashtags:
                hashtags[i] += 1
            
            else:
                hashtags[i] = 1
        # a += 1
        # if a == 100:
        #     break
    hashtags = sorted(hashtags.items(), key=lambda x: x[1], reverse=True)
    print("hashtag           hashtagCount")
    for twit in hashtags[:10]:
        print(f"\n{twit[0]}  {twit[1]}")
```

Finalmente tenemos la funcion main que llama a las demas funciones. Se puede correr para que el output entregue todos los resultados o se pueden comentar ciertos resultados para que la funcion entregue el resulatdo deseado. 

```python

def main_function(choice):
    if choice == 0: 
        print("Top 10 tweets m√°s retweeted:\n")
        top_retweet()
    elif choice == 1:
        print("Top 10 usuarios en funci√≥n a la cantidad de tweets que emitieron:\n")
        top_tweet()
    elif choice == 2:
        print("Top 10 d√≠as donde hay m√°s tweets:\n")
        top_day()
    elif choice == 3:
        print("Top 10 hashtags mas usados:\n")
        top_hashtag()
    else:
        print("Opcion ingresada incorrectamente, por favor intente nuevamente.")
choice = int(input("Ingrese 0 para los Top 10 tweets m√°s retweeted, 1 para Top 10 usuarios en funci√≥n a la cantidad de tweets que emitieron, 2 para los Top 10 d√≠as donde hay m√°s tweets y 3 para lso 10 hashtags mas usados:   "))
main_function(choice)
```
