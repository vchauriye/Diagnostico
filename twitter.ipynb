Comenzaremos por limipar la data y organizar la informacion en un dataset de pandas, con tal de que esta sea mas lejible y trabajable. Utilizaremos la metodologia que se utiliza en este lnk: https://www.kaggle.com/code/prathamsharma123/clean-raw-json-tweets-data

```python
import pandas as pd
from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings("ignore")
```

```python
raw_tweets = pd.read_json('farmers-protest-tweets-2021-03-5.json', lines=True)
raw_tweets.head(5)
```

Luego notamos que el campo de user tiene un json en vez de texto, por lo que lo normalizaremos con pandas. 

```python
users = json_normalize(raw_tweets['user'])
users.drop(['description', 'linkTcourl'], axis=1, inplace=True)
users.rename(columns={'id':'userId', 'url':'profileUrl'}, inplace=True)
users.head(5)
```

Luego quitamos los duplicados

```python
users = pd.DataFrame(users)
users.drop_duplicates(subset=['userId'], inplace=True)
print("Shape: ", users.shape)
users.head(5)
```

Ahora podemos agregarle a nuestro dataframe de tweets el campo de username

```python
user_name = []
for user in raw_tweets['user']:
    una = user['username']
    user_name.append(una)
raw_tweets['username'] = user_name
```

```python
raw_tweets.head()
```

Continuamos quitando las columnas que no son de nuestro interes y modificando la columna de date para que pase de ser tipo datetime a date.

```python
cols = [ 'id', 'username', 'date', 'renderedContent', 'retweetCount']
tweets = raw_tweets[cols]
tweets['date'] = pd.to_datetime(tweets['date']).dt.date
tweets.rename(columns={'id':'tweetId'}, inplace=True)
tweets
```

Ahora tenemos un dataframe de tweets que nos sirve para realizar los analisis pedidos.
